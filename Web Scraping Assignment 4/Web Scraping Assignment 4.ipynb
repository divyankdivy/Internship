# Q1 : Scrape the details of most viewed videos on YouTube from Wikipedia:

 Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
 You need to find following details:
 A) Rank
 B) Name
 C) Artist
 D) Upload date
 E) Views

```python
driver=webdriver.Safari()

url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
Rank = []
Name = []
Artist = []
Date = []
Views = []
```

```python
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")

Wiki = pd.DataFrame({})
Wiki['Rank'] = Rank
Wiki['Name'] = Name
Wiki['Artist'] = Artist
Wiki['Upload Date'] = Date
Wiki['Views (in Billions)'] = Views

Wiki.Name = Wiki.Name.apply(lambda x: x[:-4].strip('"'))

Wiki
```

# Q 7 : Scrape the details most watched tv series of all time from imdb.com.

  Url = https://www.imdb.com/list/ls095964455/
  You have to find the following details:
  A) Name
  B) Year span
  C) Genre
  D) Run time
  E) Ratings
  F) Votes

```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the IMDb URL
url = "https://www.imdb.com/list/ls095964455/"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Find all the TV series details
series_list = soup.find_all("div", class_="lister-item-content")

# Iterate over each TV series and extract the required details
for series in series_list:
    # Extract the name of the series
    name = series.h3.a.text.strip()
    
    # Extract the year span of the series
    year_span = series.find("span", class_="lister-item-year").text.strip("()")
    
    # Extract the genre of the series
    genre = series.find("span", class_="genre").text.strip()
    
    # Extract the run time of the series
    runtime = series.find("span", class_="runtime").text.strip()
    
    # Extract the rating of the series
    rating = series.find("span", class_="ipl-rating-star__rating").text.strip()
    
    # Extract the number of votes for the series
    votes = series.find("span", attrs={"name": "nv"})["data-value"]
    
    # Print the details of the series
    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", runtime)
    print("Rating:", rating)
    print("Votes:", votes)
    print("-----------------------------------")
```

    Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver = webdriver.Safari()
url = "https://archive.ics.uci.edu/"
driver.get(url)
```

```python
show_all_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "VIEW DATASETS")))
show_all_link.click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

#scraping Dataset name
data_name= driver.find_elements(By.XPATH,"//p[@class='normal']/b/a[1]")
for i in data_name:
    if i.text is None :
        Dataset_name.append("--") 
    else:
        Dataset_name.append(i.text)

        
#Scraping data Type
data_type=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[2]/p")
for i in data_type[1:]:
    if i.text is None :
        Data_type.append("--") 
    else:
        Data_type.append(i.text)

#Scraping Task
task=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[3]/p")
for i in task[1:]:
    if i.text is None :
        Task.append("--") 
    else:
        Task.append(i.text)
        
#Scraping Attribute_type
attribute_type=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[4]/p")
for i in attribute_type[1:]:
    if i.text is None :
        Attribute_type.append("--") 
    else:
        Attribute_type.append(i.text)
     
    #Scraping No_of_instances
no_of_instances=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[5]/p")
for i in no_of_instances[1:]:
    if i.text is None :
        No_of_instances.append("--") 
    else:
        No_of_instances.append(i.text)

        
#Scraping No_of_attribute
no_of_attribute=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[6]/p")
for i in no_of_attribute[1:]:
    if i.text is None :
        No_of_attribute.append("--") 
    else:
        No_of_attribute.append(i.text)


year=driver.find_elements(By.XPATH,"//table[@border='1']/tbody/tr/td[7]/p")
for i in year[1:]:
    if i.text is None :
        Year.append("--") 
    else:
        Year.append(i.text)
        
data_dataset=pd.DataFrame({})
data_dataset['Dataset_name']= Dataset_name
data_dataset['Data_type']= Data_type
data_dataset['Task']= Task
data_dataset['Attribute_type']= Attribute_type
data_dataset['No_of_instances']= No_of_instances
data_dataset['No_of_attribute']= No_of_attribute
data_dataset['Year']= Year
```

# Q 9 : Scrape the details of Data science recruiters from naukri.com.

  B) Designation
  C) Company
  D) Skills they hire for
  E) Location
    Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.safari.service import Service
from bs4 import BeautifulSoup

# Set up Safari WebDriver
safari_driver_path = '/usr/bin/safaridriver'
service = Service(safari_driver_path)
driver = webdriver.Safari()

# Navigate to the URL
url = "https://www.naukri.com/hr-jobs"
driver.get(url)

# Wait for the page to load
# You may need to adjust the sleep time depending on your internet speed
import time
time.sleep(5)

# Scrape the details
soup = BeautifulSoup(driver.page_source, 'html.parser')

recruiters = soup.find_all('div', {'class': 'recInfo'})
for recruiter in recruiters:
    name = recruiter.find('div', {'class': 'recName'}).text.strip()
    designation = recruiter.find('div', {'class': 'recDesign'}).text.strip()
    company = recruiter.find('div', {'class': 'recCorp'}).text.strip()
    skills = recruiter.find('div', {'class': 'recSkills'}).text.strip()
    location = recruiter.find('div', {'class': 'recLoc'}).text.strip()

    print(f"Name: {name}")
    print(f"Designation: {designation}")
    print(f"Company: {company}")
    print(f"Skills they hire for: {skills}")
    print(f"Location: {location}")
    print("---")

# Quit the browser
driver.quit()
```

```python
safari_driver_path = '/usr/bin/safaridriver'  
service = Service(safari_driver_path)
driver = webdriver.Safari()

# Navigate to the URL
url = "http://statisticstimes.com/"
driver.get(url)

# Find and click on the 'Economy' link
economy_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Economy")))
economy_link.click()

india_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, 'India')))
india_link.click()

# Find and click on the 'GDP of Indian states' link
gdp_link = driver.find_element(By.LINK_TEXT, "GDP of Indian states")
gdp_link.click()

# Wait for the page to load
# You may need to adjust the sleep time depending on your internet speed
import time
time.sleep(5)

# Scrape the details
soup = BeautifulSoup(driver.page_source, 'html.parser')

table = soup.find('table', {'id': 'table_id'})
rows = table.find_all('tr')

data = []
for row in rows[1:]:
    columns = row.find_all('td')
    rank = columns[0].text.strip()
    state = columns[1].text.strip()
    gdp_1819 = columns[2].text.strip()
    gdp_1920 = columns[3].text.strip()
    share_1819 = columns[4].text.strip()
    gdp_billion = columns[5].text.strip()

    data.append([rank, state, gdp_1819, gdp_1920, share_1819, gdp_billion])

# Create a DataFrame to store the scraped data
df = pd.DataFrame(data, columns=['Rank', 'State', 'GSDP(18-19)', 'GSDP(19-20)', 'Share(18-19)', 'GDP($ billion)'])

# Display the DataFrame
df
```

# Q 5 : Scrape the details of top 100 songs on billboard.com.

     Url = https://www.billboard.com/
     You have to find the following details:
     A) Song name
     B) Artist name
     C) Last week rank
     D) Peak rank
     E) Weeks on board
      Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
# Send a GET request to the Billboard website
url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the container that holds the song details
songs_container = soup.find('ol', class_='chart-list__elements')

# Check if the songs_container is found
if songs_container is not None:
    # Initialize lists to store song details
    song_names = []
    artist_names = []
    last_week_ranks = []
    peak_ranks = []
    weeks_on_board = []

    # Loop through each song entry in the container and extract the details
    for song in songs_container.find_all('li', class_='chart-list__element'):
        song_name = song.find('span', class_='chart-element__information__song').get_text()
        artist_name = song.find('span', class_='chart-element__information__artist').get_text()
        last_week_rank = song.find('span', class_='chart-element__meta text--last').get_text()
        peak_rank = song.find('span', class_='chart-element__meta text--peak').get_text()
        weeks_on_chart = song.find('span', class_='chart-element__meta text--week').get_text()

        song_names.append(song_name)
        artist_names.append(artist_name)
        last_week_ranks.append(last_week_rank)
        peak_ranks.append(peak_rank)
        weeks_on_board.append(weeks_on_chart)

    # Create a DataFrame to store the scraped data
    df = pd.DataFrame({
        'Song Name': song_names,
        'Artist Name': artist_names,
        'Last Week Rank': last_week_ranks,
        'Peak Rank': peak_ranks,
        'Weeks on Board': weeks_on_board
    })

    # Display the DataFrame
    print(df)
else:
    print("Failed to find the songs container on the website. Please check the URL and website structure.")
```

# Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description C) Contributors count
D) Language used

```python
driver = webdriver.Safari()  # You can use another driver if you prefer

# Navigate to the GitHub website
url = 'https://github.com/'
driver.get(url)

# Click on the Explore menu
explore_menu = driver.find_element(By.LINK_TEXT, 'Explore')
explore_menu.click()

# Click on the Trending option from the Explore menu
trending_option = driver.find_element(By.XPATH, "//a[@href='/trending']")
trending_option.click()

# Wait for the trending repositories to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'ol.repo-list')))

# Parse the page source using BeautifulSoup
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find the container that holds the trending repositories
repo_list = soup.find_all('li', class_='repo-list-item')

# Initialize lists to store repository details
repo_titles = []
repo_descriptions = []
contributors_counts = []
languages_used = []

# Loop through each repository entry and extract the details
for repo in repo_list:
    # Extract the title
    repo_title = repo.find('h1').text.strip()
    repo_titles.append(repo_title)

    # Extract the description (if available)
    repo_desc = repo.find('p', class_='mb-1').text.strip() if repo.find('p', class_='mb-1') else "No description available"
    repo_descriptions.append(repo_desc)

    # Extract the contributors count
    contributors_count = repo.find('a', href=lambda href: href and '/contributors' in href).text.strip()
    contributors_counts.append(contributors_count)

    # Extract the language used (if available)
    language_used = repo.find('span', itemprop='programmingLanguage').text.strip() if repo.find('span', itemprop='programmingLanguage') else "Not specified"
    languages_used.append(language_used)

# Create a DataFrame to store the scraped data
df = pd.DataFrame({
    'Repository Title': repo_titles,
    'Repository Description': repo_descriptions,
    'Contributors Count': contributors_counts,
    'Language Used': languages_used
})

# Close the web driver
driver.quit()

# Display the DataFrame
print(df)
```
